## Skeleton for plotting GIFs

First import the libraries you will need to use in this notebook. 

```python
import numpy as np
import matplotlib.pyplot as plt
import imageio
import os
```

### Generate and example plot
Using matplotlib, generate one frame of you GIF

```python
x = np.linspace(0,15,100)
a = 1
y = a*np.cos(x)
y2 = np.sin(a*x)
```

```python
fig, ax = plt.subplots()
ax.plot(x,y, label='cos(x)')
ax.plot(x,y2, label='sin(x)')
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.legend(frameon=False)
plt.show()
```

### Test your plot over time
Try iterating through a couple values in time to make sure your plots are behaving as expected

```python
a = np.linspace(0,2,5)
for i in a:
    fig, ax = plt.subplots()
    y = i*np.cos(x)
    y2 = np.sin(x*i)
    ax.plot(x,y, label=f'{i}cos(x)')
    ax.plot(x,y2, label=f'sin({i}x)')
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.legend(frameon=False)
plt.show()
```

### Generate you pngs
Increase the number of steps to your desired size and generate your plots in a loop. Each iteration saves your plot in the directory along with it's assigned name in a list called filenames. 

```python
filenames = []
a = np.linspace(0, 2, 50)
for i in a:
    fig, ax = plt.subplots()
    y = i*np.cos(x)
    y2 = np.sin(x*i)
    ax.plot(x,y, label=f'{i:2f}cos(x)')
    ax.plot(x,y2, label=f'sin({i:2f}x)')
    ax.spines['right'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.legend(frameon=False, loc = 'lower right')
    filename = f'{i}.png'
    filenames.append(filename)
    plt.savefig(filename,dpi=100)
    plt.close()
```

### Stitch the pngs into a GIF!
This will iterate through filenames using os and stitch them into GIF using imageio. Don't worry, the pngs will be removed from your directy and you will just be left with a GIF

```python
with imageio.get_writer('curvy.gif', mode='I') as writer:
    for filename in filenames:
        image = imageio.imread(filename)
        writer.append_data(image)
        
# Remove files
for filename in set(filenames):
    os.remove(filename)
```
